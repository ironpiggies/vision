```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv.imread("test2_Color.png")
width = 10.
```

```python
img = cv.resize(img, (640, 360))
```

```python
def showimg(image, cmap="gray"):
    plt.figure(figsize=(width, width*img.shape[0]/img.shape[1]))
    RGB_img = cv.cvtColor(image, cv.COLOR_BGR2RGB)
    plt.imshow(RGB_img, cmap=cmap, aspect='auto')
    plt.show()
showimg(img)
```

```python
blur = cv.GaussianBlur(img, (7, 7), 0)
showimg(blur)
```

```python
# gray = cv.cvtColor(blur, cv.COLOR_BGR2GRAY)
# showimg(gray)
```

```python
edges = cv.Canny(blur, 50, 150, apertureSize = 3)
showimg(edges)
```

```python
circles = cv.HoughCircles(edges, cv.HOUGH_GRADIENT, 1, 5, param1=40, param2=25, minRadius=0, maxRadius=70)
img_copy = np.copy(img)
if circles is not None:
    circles = np.round(circles[0, :]).astype("int")
    
    for (x, y, r) in circles:
        if r < 10:
            cv.circle(img_copy, (x, y), r, (0, 255, 0), 4)
        if r > 10 and r < 14:
            cv.circle(img_copy, (x, y), r, (255, 0, 0), 4)
        if r > 65 and r < 68:
            cv.circle(img_copy, (x, y), r, (0, 0, 255), 4)

showimg(img_copy)
```
